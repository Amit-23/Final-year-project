```python
import pandas as pd
import requests
from datetime import datetime
import whois
from urllib.parse import urlparse

def get_page_rank(url: str):
    headers = {
        'API-OPR': 'go48o448ws8cok8k88sk4kc0wcogowoggg4gccko'
    }
    domain = urlparse(url).netloc
    response = requests.get(f'https://openpagerank.com/api/v1.0/getPageRank?domains%5B0%5D={domain}', headers=headers)
    if response.status_code == 200:
        data = response.json()
        if data and 'status_code' in data and data['status_code'] == 200 and data['response']:
            return data['response'][0]['rank']
    return None

#def get_domain_expiration(url: str):
 #   domain = urlparse(url).netloc
  #  domain = whois.query(domain).expiration_date
   # if domain:
    #    return domain
    #return None

def get_expanded_urls(csv_filename: str, field_name: str = 'url'):
    urls = pd.read_csv(csv_filename)
    urls["expanded_url"] = ""
    urls["inner_html"] = ""
    urls["http_status"] = ""
    urls["is_@_present"] = False
    urls["is___present"] = False
    urls["is_#_present"] = False
    urls["is_redirected"] = False
    urls["number_of_redirects"] = 0
    urls["error_fetch"] = False
    urls["page_rank"] = None
    urls["is_abnormal"] = False
    #urls["domain_expiration"] = None
    urls = urls.reset_index()
    count = urls.shape[0]
    for index, row in urls.iterrows():
        if index > 100: continue
        print(f"{index}/{count}")
        try:
            expand_response = requests.get(row[field_name])
            if expand_response.history:
                urls.loc[index, "is_redirected"] = True
                urls.loc[index, "number_of_redirects"] = len(expand_response.history)
                urls.loc[index, "http_status"] = expand_response.status_code
                if '@' in expand_response.url:
                    urls.loc[index, "is_@_present"] = True
                if '_' in expand_response.url:
                    urls.loc[index, "is___present"] = True
                if '#' in expand_response.url:
                    urls.loc[index, "is_#_present"] = True
                urls.loc[index, "expanded_url"] = expand_response.url
                urls.loc[index, "inner_html"] = expand_response.content
            else:
                urls.loc[index, "http_status"] = expand_response.status_code
                urls.loc[index, "expanded_url"] = expand_response.url
                urls.loc[index, "inner_html"] = expand_response.content

            page_rank = get_page_rank(expand_response.url)
            if page_rank is not None:
                urls.loc[index, "page_rank"] = page_rank

        except Exception as e:
            print(e)
            urls.loc[index, "error_fetch"] = True

    urls.to_csv("output.csv", index=False)
    
if __name__ == '__main__':
    get_expanded_urls('test.csv', field_name='url')
```
